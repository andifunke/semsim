Measures correlation of multiple similarity and relatedness metrics with human similarity and associativeness scores for German word pairs. Metrics are described in https://github.com/andifunke/topic-labeling and [Topic Labeling](https://drive.google.com/file/d/1IMX4RJMAiJ61RLZ_ZHAnMwc7EB7-gv5f).

```python
import pandas as pd

from semsim.constants import DATA_DIR
```

```python
file = DATA_DIR / "metrics" / "Rating_word_pairs_topic-scores.csv"
df = pd.read_csv(file, index_col=0).set_index(['target', 'prime'])
df
```

```python
corr = df.corr('pearson')
corr
```

```python
corr_file = DATA_DIR / "metrics" / "Rating_word_pairs_topic-scores_correlations.csv"
corr[['ass_val', 'sim_val']].to_csv(corr_file, float_format='%.03f')
corr[['ass_val', 'sim_val']]
```
