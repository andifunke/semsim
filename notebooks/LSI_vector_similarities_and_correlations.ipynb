```python
file_gtp = "../data/SemD/V_lsi_document_vectors.csv"
file_skl = "./data/SemD/V_lsi_sklearn_document_vectors.csv"
file_skl_cent = "../data/SemD/V_lsi_sklearn_cent_document_vectors.csv"

df_gtp = pd.read_csv(file_gtp, index_col=0)
df_skl = pd.read_csv(file_skl, index_col=0)
```

```python
df_gtp.shape, df_skl.shape
```

```python
def avg_corr(vector, others, samples=10):
    vectors = others.sample(10, replace=False)
    correlations = vectors.apply(lambda x: pearsonr(x.values, vector.values)[0], axis=1)
    mean_corr = correlations.mean()
    return mean_corr

df_gtp[:1000].apply(lambda x: avg_corr(x, df_gtp), axis=1).mean()
```

```python
df_skl[:1000].apply(lambda x: avg_corr(x, df_skl), axis=1).mean()
```

```python
def avg_sim(vector, others, samples=10):
    vectors = others.sample(10, replace=False)
    similarities = vectors.apply(lambda x: 1 - spatial.distance.cosine(vector, x), axis=1)
    mean_sim = similarities.mean()
    return mean_sim

df_gtp[:1000].apply(lambda x: avg_sim(x, df_gtp), axis=1).mean()
```

```python
df_skl[:1000].apply(lambda x: avg_sim(x, df_skl), axis=1).mean()
```
